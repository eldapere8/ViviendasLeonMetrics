# ViviendasLeón Newsletter Analysis

### Executive Summary

1. Users open ViviendasLeón's newsletters at a higher-than-average rate. The open rate was higher for results-oriented subject lines such as 'ViviendasLeón News Update: Kitchen Completed!', which we recommend continuing in the future.
2. Recipients clicked on emails at a lower-than-average rate (1.1%). We recommend trying to boost this metric with strategies such as personalizing emails, experimenting with different send times and A/B testing features.
3. There are a higher number of unsubscriptions than expected. In order to lower this number, we recommend avoiding excessive emailing, improving email design and providing more value to the user in the form of blog articles, discounts, etc.
4. The bounce rate and delivery rate are within industry standards. This means ViviendasLeón has a proper subscriber list and most of the emails being sent are indeed being received. 
5. The top performing newsletters contain colorful high quality photos, shorter text/subject lines, and results-oriented titles and subject lines. 

### Data and Metrics
The data below comes from a Mailchimp csv export containing email campaign data between January 2019 and May 2020. <br>
The top metrics we will use to measure newsletter campaign performance are:
<br>
1. Bounce Rate 
2. Delivery Rate 
3. Open Rate 
4. Click Rate
5. Unsubscribe Rate.

Source: https://writtent.com/blog/10-metrics-to-track-your-newsletter-content-success/

### Quick Imports and Data Cleaning

```python
# These lines import the Numpy, Pandas and Datascience modules.
import numpy as np
import pandas as pd
import datetime

# These lines do some fancy plotting magic.
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
first_newsletter = pd.read_csv("VL Social Media Data 1_1_19 - 6_16_20 - MailChimp Info.csv")
second_newsletter = first_newsletter[first_newsletter["Emails sent"] > 10]
```

```python
second_newsletter["Content type"].unique()
newsletter = second_newsletter.drop(['Content type'], axis = 1)
```

### 1. Bounce Rate 

Bounce Rate: the total percentage of your emails that weren’t delivered successfully.

```python
#Code to plot bounce rate over time.

plt.figure(figsize=(10,5))
plt.plot(newsletter["Date"], round(100*newsletter["Bounces"]/newsletter["Emails delivered"], 2))
plt.title("Bounce Rate over Time")
plt.ylabel("Bounce Rate (%)")
plt.xticks(rotation=60)
plt.show()

print("The total number of emails that weren't delivered successfully was " + str(newsletter["Bounces"].sum()) + ".")
print("The percentage of total emails that weren't delivered successfully was " + str(round(100*newsletter["Bounces"].sum()/newsletter["Emails delivered"].sum(), 2)) + "%.")
print("Since the bounces constitute less than 2% of total emails delivered, it is not significant enough to act on.")

```

### 2. Delivery Rate 

Delivery Rate: the percentage of messages which ended up in your contacts’ inboxes. <br>
Formula: Total Emails Delivered / Total Emails Sent = Delivery Rate <br>
The delivery rate should ideally be over 95%.

```python
#Code to plot delivery rate over time.

plt.figure(figsize=(10,5))
plt.plot(newsletter["Date"], round(100*newsletter["Emails delivered"]/newsletter["Emails sent"], 2))
plt.title("Delivery Rate over Time")
plt.ylabel("Delivery Rate (%)")
plt.xticks(rotation=60)
plt.show()

print("The delivery rate is " + str(round(100*newsletter["Emails delivered"].sum()/newsletter["Emails sent"].sum(), 2)) + "%.")
print("Since the delivery rate is above 95%, it is not significant enough to act on.")
```

### 3. Open Rate 

Open Rate: the percentage of total email contacts who open your email communications as opposed to deleting the message or banishing it to the depths of their inbox.
The open rate is typically in the 15-20% range.


```python
#Code to plot open rate over time. 

plt.figure(figsize=(10,5))
plt.plot(newsletter["Date"], newsletter["Open rate"])
plt.title("Open Rate over Time")
plt.ylabel("Open Rate (%)")
plt.xticks(rotation=60)
plt.show()

print("The average open rate is " + str(round(newsletter["Open rate"].mean(), 2)) + "%.")
print("This is above the typical rate. Overall, ViviendasLeon has a good open rate.")
print("The latest spike in 12/7/19 was likely to due to the subject line 'ViviendasLeón News Update: Kitchen Completed!'.")
print("We recommend similar result-oriented subject lines in the future.")
```

### 4. Click Rate

Click Rate: the percentage of delivered emails that were engaged with at least once. <br>
The nonprofit click rate average is 2.0%.

```python
#Code to plot click rate over time.

plt.figure(figsize=(10,5))
plt.plot(newsletter["Date"], newsletter["Click rate"])
plt.title("Click Rate over Time")
plt.ylabel("Click Rate (%)")
plt.xticks(rotation=60)
plt.show()

print("The average click rate is " + str(round(newsletter["Click rate"].mean(), 2)) + "%.")
print("This is below the nonprofit average. We recommend attempting strategies to boost click rates.")
print("This would include personalizing emails, experimenting with different send times and A/B testing features.")

print("Note: The spike seen on 5/15 is a result of a low number of recipients.")
```

### 5. Unsubscribe Rate 

Unsubscribe Rate: the percentage of people who received your communications who chose to opt out of receiving newsletter content from you in the future. <br>
The nonprofit average unsubscribe rate is 0.10%. 

```python
#Code to plot unsubscribe rate over time.

plt.figure(figsize=(10,5))
plt.plot(newsletter["Date"], round(100*newsletter["Unsubscribes"]/newsletter["Emails delivered"], 2))
plt.title("Unsubscribe Rate over Time")
plt.ylabel("Unsubscribe Rate (%)")
plt.xticks(rotation=60)
plt.show()

print("The average unsubscribe rate is " + 
      str(round(100*newsletter["Unsubscribes"].sum()/newsletter["Emails delivered"].sum(), 2)) + "%.")
print("This is above the nonprofit average of 0.1%. We recommend attempting strategies to lower unsubscribe rates.")
print("Ways to improve this metric could be avoiding excessive emailing, improving email design and providing value " + 
     "to the user in the form of blog articles, discounts, etc.")

print("Note: The spike in unsubscribes on 5/23 are likely due to high email frequency.")
```

### Additional Analysis: Subject Line Length

The effect of subject line length on open rate. It appears they have a slight negative correlation, meaning that the shorter the subject line, the more people will open the email.

```python
plt.figure(figsize=(10,5))
plt.scatter(newsletter["Subject line"].str.len(), newsletter['Open rate'])
plt.title("Subject Line Length vs Open Rate")
plt.xlabel("Subject Line Character Length")
plt.ylabel("Open Rate")
plt.show()
```
